```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import requests
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import csv
import pandas as pd

#  Any infinity scroll URL
mainURL = "https://www.amazon.ca/"
var = "shoes"

# url = "https://in.pinterest.com/search/pins/?q=" + var 

url = "https://www.amazon.ca/s?k=shoes&i=fashion&rh=n%3A21204935011%2Cp_89%3ANew+Balance%7CPUMA%7CReebok%7CUnder+Armour%7Cadidas%7Cadidas+Originals&dc&crid=1FM4ZC7WXIZXY&qid=1679116249&rnid=7590290011&sprefix=shoes%2Caps%2C117&ref=sr_nr_p_89_7&ds=v1%3Ai5w7lVGDYWyb82%2Fi49XTrwzHT0dI9k8ljFGljWPdMho" 
print("URL",url);

sleepTimer = 3    # Waiting 1 second for page to load
image_limits = 6000
images_per_page = []
no_of_images = []

images = []
title = []
price = []
rating= []
brand = []
review_count = []

#  Bluetooth bug circumnavigate
options = webdriver.ChromeOptions() 
options.add_experimental_option("excludeSwitches", ["enable-logging"])

driver = webdriver.Chrome("C:/Program Files (x86)/chromedriver.exe", options=options)  # path=r'to/chromedriver.exe'
driver.get(url)

csv_data = []

def scroolWindow(): 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)");
    time.sleep(sleepTimer)
    scrapeImages()
    
def scrapeImages():
    soup = BeautifulSoup(driver.page_source,'html.parser')
    for (i, divDom) in enumerate(soup.find_all('div', {'class': 'a-spacing-base'})):

        if divDom.find('h2') and divDom.find('span', {'class': 'a-offscreen'}) and divDom.find('span',{'class':'a-size-base s-underline-text'}) and divDom.find('span',{'class':'a-size-base'}) and divDom.find('h5', {'class': 's-line-clamp-1'}) and not divDom.find('div', {'class': 'sbv-product-container'}):
            if divDom.find('h2').text not in title:
                title.append(divDom.find('h2').text);
                price.append(divDom.find('span', {'class': 'a-offscreen'}).text)
                rating.append(divDom.find('span', {'class': 'a-size-base'}).text)
                brand.append(divDom.find('h5', {'class': 's-line-clamp-1'}).text)
                review_count.append(divDom.find('span', {'class': 'a-size-base s-underline-text'}).text)
                '''for (i, link) in enumerate(divDom.find_all('img')):
                    no_of_images.append(link.get('src'))
#                     images_per_page.append(link.get('src'))
                    img_data = requests.get(link.get('src')).content
                    with open(f'{len(no_of_images)} image.jpg', 'wb') as handler:
                        csv_data.append([(len(csv_data) + 1), f'{len(no_of_images)} image.jpg', divDom.find('h2').text, divDom.find('span', {'class': 'a-offscreen'}).text])
                        handler.write(img_data)'''
     
    if len(rating) < image_limits:
        try:
            next_button = driver.find_element(By.XPATH, "//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
            next_button.click()
            scroolWindow()
        except NoSuchElementException:
            # Handle the exception
            # Save the scraped data to a CSV file
            print("IN NoSuchElementException")
            '''with open('amazon scraped '+var+' data.csv', 'w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(['ID', 'Title', 'Price','Rating','Brand','review_count'])
                for row in csv_data:
                    writer.writerow(row)'''
        finally:
            # Quit the webdriver instance
            print("IN FINALLY")
            with open('amazon scraped '+var+' data.csv', 'w', newline='') as file:
                writer = csv.writer(file)
                data=pd.DataFrame()
                data["Company"]=pd.Series(brand)
                data["Title"]=pd.Series(title)
                data["Price"]=pd.Series(price)
                data["Rating"]=pd.Series(rating)
                data["review_count"]=pd.Series(review_count)

                #data["Image"]=pd.Series(Image)
                gfg_csv_data = data.to_csv('C:/Users/ACER/Downloads/shoe_meta6.csv', index = True)
                print('\nCSV String:\n', gfg_csv_data)
            driver.quit()

scrapeImages()
```
